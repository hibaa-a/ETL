#  Create new data with select columns

```python
new_data_acc = data_acc[['Accident_Index', 'Accident_Severity','Day_of_Week','Weather_Conditions']].copy()
new_data_acc.head()
```

```python
new_data_veh = data_veh[['Accident_Index', 'Age_Band_of_Driver','Age_of_Vehicle','make','model','Sex_of_Driver']].copy() 
new_data_veh.head()
```

# Change the column names and drop NaN, data missing/ out of range values. 

```python
accident = new_data_acc.rename(
    columns = {"Accident_Index": "id",
                "Accident_Severity": "severity",
                "Day_of_Week": "day",
                "Weather_Conditions": "weather"
    }
)
```

```python
vehicle0 = new_data_veh.rename(
    columns = {"Accident_Index": "id",
                "Age_Band_of_Driver": "driver_age",
                "Age_of_Vehicle": "vehicle_age",
                "make": "make",
                "model": "model",
                "Sex_of_Driver": "sex"
    }
)
```

```python
vehicle0
```

```python
vehicle0.sex.unique()
```

```python
df_vehicle = vehicle0[vehicle0.sex != 'Data missing or out of range']
```

```python
df_vehicle1 = df_vehicle[df_vehicle.sex != 'Not known']
```

```python
df_vehicle1.driver_age.unique()
```
